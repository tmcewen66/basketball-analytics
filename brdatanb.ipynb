```python
conn = sqlite3.connect('nba_stats.db')

query = '''
SELECT name, scoring_plus, per_100_pts, true_shooting_percentage, pts_plus, ts_plus
FROM scoring_plus 
WHERE season_end_year = 2026
ORDER BY scoring_plus DESC
'''
scoring_plus_2026_df = pd.read_sql_query(query, conn)
conn.close()
scoring_plus_2026_df.head(25)
```

```python
conn = sqlite3.connect('nba_stats.db')

query = '''
SELECT s.name, s.season_end_year, s.slug, s.scoring_plus, s.per_100_pts, s.true_shooting_percentage, s.pts_plus, s.ts_plus
FROM scoring_plus as s
JOIN basic_stats as b ON s.slug = b.slug AND s.season_end_year = b.season_end_year
WHERE b.minutes_played > 1500
ORDER BY scoring_plus DESC
'''
scoring_plus_leaders_df = pd.read_sql_query(query, conn)
conn.close()
scoring_plus_leaders_df.head(25)
```

```python
conn = sqlite3.connect('nba_stats.db')

query = 'SELECT * FROM player_profile WHERE qualified = 1'
player_profile_df = pd.read_sql_query(query, conn)
conn.close()

player_profile_df
```

```python
plus_leaders_2026 = player_profile_df[player_profile_df['season_end_year'] == 2026].sort_values(by='scoring_plus', ascending=False).head(25)
plus_leaders_2026
```

```python
ppg_leaders_2026 = player_profile_df[player_profile_df['season_end_year'] == 2026].sort_values(by='points_per_game', ascending=False).head(25)
ppg_leaders_2026
```
